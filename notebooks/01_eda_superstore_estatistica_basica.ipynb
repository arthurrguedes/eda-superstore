```python
# Importando as bibliotecas padrão
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
# Carregando o dataset
df = pd.read_csv("superstore.csv")
```

```python
df.head()
df.info()
```

```python
df.dtypes
```

O dataset possui variáveis numéricas como 'Sales', 'Profit' e 'Quantity', além de variáveis categóricas como 'Category', 'Segment' e 'Region'.

```python
df.describe()
```

A variável 'Sales' mostra uma assimetria forte. Embora a média seja próxima de 230, a mediana é bem menor (54), o que indicaque a maior parte das vendas são de valores baixos. 75% das vendas não ultrapassam de 210, enquanto existem valores máximos acima de 20 mil. Dessa forma, a mediana se mostra uma medida mais aasertiva do comportamento das vendas do que a média.

As variáveis 'Row ID' e 'Postal Code' não possuem significado estatístico direto para análise
descritiva, sendo utilizadas apenas como identificadores ou categorias.


```python
df["Sales"].mean(), df["Sales"].median()
```

```python
# Gráfico Histograma - Distribuição das Vendas por Frequência
plt.figure(figsize=(8,5))
sns.histplot(df["Sales"], bins=30, kde=True)
plt.title("Distribuição das Vendas (Sales)")
plt.xlabel("Valor da Venda")
plt.ylabel("Frequência")
plt.show()
```

O histograma confirma a análise feita no 'describe' anteriormente, evidenciando uma concentração de vendas em valores baixos e uma cauda longa a direita, característica de uma distribuição assimétrica positiva.



```python
# Gráfico Boxplot das Vendas
# Mediana, Quartis e Outliers
plt.figure(figsize=(8,5))
sns.boxplot(x=df["Sales"])
plt.title("Boxplot das Vendas (Sales)")
plt.xlabel("Valor da Venda")
plt.show()
```

O boxplot reforça a presença de diversos outliers à direita, indicando que a média é
fortemente influenciada por valores extremos e não representa adequadamente o
comportamento típico das vendas. Nesse contexto, a mediana se mostra uma medida mais
robusta e adequada para análises descritivas.

```python
# Gráfico Boxplot de Vendas por Categoria
# Qual categoria tem vendas mais previsíveis?
# Qual tem mais extremos?
plt.figure(figsize=(8,5))
sns.boxplot(x="Category", y="Sales", data=df)
plt.title("Distribuição de Vendas por Categoria")
plt.xlabel("Categoria")
plt.ylabel("Sales")
plt.show()
```

Na análise por categoria, observa-se que a categoria Furniture apresenta vendas mais
previsíveis, com menor dispersão, enquanto Technology concentra os maiores valores e
outliers mais extremos, sendo responsável pelas vendas mais altas da loja

Insight de negócio:
Estratégias comerciais devem considerar que poucas vendas de alto valor são responsáveis por grande parte do faturamento, especialmente na categoria Technology.
